# Abstract Data Type (ADT)

## Máquina de Números Naturais

```python
def create_natural():
    current = -1

    def next():
        nonlocal current
        current += 1
        return current

    return next
```

````guile
(define (create-natural)
  (let ((current -1))
    (lambda ()
      (set! current (+ current 1))
      current)))
````

```python

next_natural2 = create_natural()

```
