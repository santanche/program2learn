## Exercício A

Dada uma estrutura do tipo:

```guile
(define friends-list
  '((Asdrúbal 25)
    (Doriana 30)
    (Quincas 28)))
```

Escreva uma função que receba a estrutura e gere uma lista apenas com o segundo elemento (lista de idades).

```guile
(define (ages people-list)
  (if (null? people-list)
      '()
      (cons (cadr (car people-list))
            (ages (cdr people-list)))))

(ages friends-list)
```

# Função de Ordem Superior

## Exercício B

Quebre a função anterior em duas funções:
* função `second`: recebe uma lista com dois elementos e retorna o segundo elemento
* função `process`: recebe dois parâmetros -- 

```guile
(define (second item)
  (cadr item))

(define (process items-list func)
  (if (null? items-list)
      '()
      (cons (func (car items-list))
            (process (cdr items-list) func))))

(process friends-list second)
```

# `map` - função de ordem superior

```guile
(map second friends-list)
```
