# Estrutura com Lista

````guile
(define pessoa
  '((Asdrúbal 25)
    (Doriana 30)
    (Quincas 28)))
````

````guile
(car pessoa)
````

````guile
(cdr pessoa)
````

````guile
(car (cdr pessoa))
````

````guile
(cdr (car (cdr pessoa)))
````

````guile
(car (cdr (car (cdr pessoa))))
````

# Abreviações `cxr`

````guile
(cadr pessoa)
````

````guile
(cadr (cadr pessoa))
````
