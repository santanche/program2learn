````guile
'( x . () )
````

# Criando Primitivas SQL-like

## Criando um Lista de Associação

A lista representa o esquema de um conjunto de dados.

````guile
(define prehistoric-schema
  '((name . 0) (type . 1) (period . 2) (size . 3) (diet . 4)))
````

## `assq`

````guile
(assq 'type prehistoric-schema)
````

````guile
(cdr (assq 'period prehistoric-schema))
````

## `list-ref`

````guile
(define prehistoric-record '("Tyrannosaurus Rex" dinosaur Cretaceous 12.0 carnivore))
````

````guile
(list-ref prehistoric-record 2)
````

## Usando `iota` para produzir pares `(coluna . posição)` automaticamente

````guile
(iota (length '(name type period length diet)))
````

````guile
(define (indexed-pairs lst)
  (map cons lst (iota (length lst))))
````

````guile
(indexed-pairs '(name type period length diet))
````
